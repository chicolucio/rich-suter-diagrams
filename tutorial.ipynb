Rich and Suter diagrams are a very useful tool to explain the electron configurations of all transition elements, and in particular, the $s^1$ and $s^0$ configurations of the elements Cr, Cu, Nb, Mo, Ru, Rh, Pd, Ag, and Pt. They were proposed by Rich and Suter in Rich, R. L.; Suter, R. W.; J. Chem. Educ. 1988, 65, 702. [Link](https://pubs.acs.org/doi/pdf/10.1021/ed065p702)
